**Objective**

**Problem Statement**

Classification is an important task in machine learning. In this project, we build a Decision Tree model to classify data into different categories using rule-based logic.

Example problem:
Predict whether a person will buy a product or not based on features such as age and income.

**Dataset Description**

Example dataset:

| Age   | Income | Student | Buy Product |

# **Step 1: Import Libraries**

```
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

# **Step 2: Create Dataset**

```
data = {
    'Age': ['<30','<30','31-40','>40','>40','>40','31-40','<30'],
    'Income': ['High','High','High','Medium','Low','Low','Low','Medium'],
    'Student': ['No','No','No','No','Yes','Yes','Yes','No'],
    'Buy': ['No','No','Yes','Yes','Yes','No','Yes','No']
}

df = pd.DataFrame(data)
print(df)
```

# **Step 3: Convert Categorical Data to Numerical**

```
le = LabelEncoder()

for column in df.columns:
    df[column] = le.fit_transform(df[column])

print(df)
```

# **Step 4: Split Data into Features and Target**

```
X = df.drop('Buy', axis=1)
y = df['Buy']
```

# **Step 5: Train Decision Tree Model**

```
model = DecisionTreeClassifier()
model.fit(X, y)
```

# **Step 6: Make Predictions**

```
prediction = model.predict(X)
print("Predictions:", prediction)
```

# **Step 7: Visualize the Decision Tree**

```
plt.figure(figsize=(10,6))
plot_tree(model, feature_names=X.columns, class_names=['No','Yes'], filled=True)
plt.show()
```

# **Generated Decision Rules (Example)**

The decision tree may generate rules such as:

IF Age < 30 AND Income = High
THEN Buy = No

IF Age = 31–40
THEN Buy = Yes

IF Age > 40 AND Student = Yes
THEN Buy = Yes

These rules are automatically learned from the dataset.
